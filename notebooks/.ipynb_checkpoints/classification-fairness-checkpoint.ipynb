# Models

```python
def train_model(layer_0, layer_1):
#Trains a tabular model model with 2 layers of sizes layer_0 and layer_1 and returns its accuracy
    n_epochs = 6    
    learn = tabular_learner(dls, layers=[int(layer_0), int(layer_1)])
    with learn.no_logging():
        lr = learn.lr_find(show_plot=False)[0]
        learn.fit_one_cycle(n_epochs, lr)
    preds, targs = learn.get_preds()
    preds_raw = torch.log(preds/(1-preds)) # inverse_of_sigmoid
    return accuracy(preds_raw, targs).numpy()

def get_predictions(df, learn):
    df_dls = learn.dls.valid_ds.new(df)
    df_dls.process()
    df_dls.conts = df_dls.conts.astype(np.float32)
    dl = learn.dls.valid.new(df_dls)
    #################################################################################
    probs, classes = learn.get_preds(dl=dl, with_input=False, with_decoded=False)
    results = pd.DataFrame(columns=['p0', 'p1', 'class'], index=range(len(classes)))
    results.loc[:, ['p0', 'p1']] = probs.numpy()
    results.loc[:, 'class'] = classes.numpy()
    return results
```

## Bayesian Modeling

```python
def get_predictions_pyro(model, guide, df, n_samples=25, varbls=['_RETURN']):
    
    df_dls = learn.dls.valid_ds.new(df)
    df_dls.process()
    df_dls.conts = df_dls.conts.astype(np.float32)
    dl = learn.dls.valid.new(df_dls)
    x_num = torch.Tensor(dl.dataset.conts.values).to(device, torch.float32)
    x_cat = torch.LongTensor(dl.dataset.cats.values).to(device)
    
    predictive = Predictive(model, guide=guide, num_samples=n_samples, return_sites=varbls)
    samples_post = {v: predictive(x_cat, x_num, None)[v].cpu().detach().numpy() for v in varbls}
    
    return samples_post
```

### Bayesian NN

#### Model definition 

---------------------------------
Let $P(\cdot | \mathbf{w})$ be the neural network architecture define above with weight parameters $\mathbf{w}$ i.e. $P(x| \mathbf{w})$ is the probability that observation $x$ is in the class 1 (person $x$ has income greater than 50k). Now let $w^*$ be the weights found by the training process using fastai. We define our bayesian model for the class type $y$ with these equations:


$$\begin{align*}  y &\sim \text{Bernoulli}(P(x | \mathbf{w})) \\
\mathbf{w} &\sim \text{Normal}(w^*, \sigma^2)
\end{align*} 
$$

With this definition we have that our non-normalized log-posterior density for the weights $\mathbf{w}$ is:

$$
\begin{align*}
L(\mathbf{w}) &= \sum_{i=1}^N y_i \log P(x_i | \mathbf{w}) + (1-y_i)\log \left( 1 - P(x_i | \mathbf{w})\right) -  \frac{1}{2} \left[(\mathbf{w}-w^*)^{\rm T}\Sigma^{-1}(\mathbf{w}-w^*) + \log |\Sigma|\ +  d\log(2\pi) \right]
\\
& = - \left( \sum_{i=1}^N \text{CrossEntropy}(y_i, P(x_i | \mathbf{w})) +  \sum_{j=1}^d  \left[\frac{1}{2\sigma_j^2}(\mathbf{w}_j - w^*_j)^2 + \log \sigma_j \right] + \log\left[\left( 2\pi \right)^{\frac{d}{2}}\right] \right)
\end{align*} 
$$
where $\Sigma$ is the covariance matrix (in this case a diagonal matrix containing all the individual variances) and $d$ is the number of weights.

So in this case we have that our log-posterior density is a cross entropy with regularization that penalized the weights to be close to the ones trained by fastai

We extract tensors of numerical and categorical variables on the training set

```python
device = learn.dl.device

x_num_train = torch.Tensor(learn.dls.dataset.conts.values).to(device, torch.float32)
x_cat_train = torch.LongTensor(learn.dls.dataset.cats.values).to(device)
y_train = torch.Tensor(learn.dls.dataset.y.values).to(device)
x_num_train, x_cat_train, y_train
```

```python
batch_size = 1000
scale = 0.001

fastai_centered = True

#original
def model(x_cat, x_num, y=None):
    
    #define priors
    priors = {}
    for name, values in learn.model.named_parameters():
        if fastai_centered:
            loc = values
        else:
            loc = torch.zeros_like(values)
            
        priors[name] = dist.Normal(loc=loc, 
                                   scale=scale*torch.ones_like(values))
    
    lifted_module = pyro.random_module("module", learn.model, priors)
    lifted_module_sample = lifted_module()
    
    probs = softmax(lifted_module_sample(x_cat, x_num), dim=1)[:, 1] #probabilidad clase 1
    
    #likelihood:
    if y is not None:
        with pyro.plate('observe_data', size=len(y), subsample_size=batch_size, device=device) as ind:
            pyro.sample('obs', dist.Bernoulli(probs.index_select(0, ind)), obs=y.index_select(0, ind))
    
    return probs
```

Every time we execute the following cell we are sampling from our prior distribution

```python
model(x_cat_train, x_num_train, y_train)
```

```python
def show_prior(samples):
    persons_by_class = samples.groupby('y').person_id.unique().to_dict()
    person_0 = np.random.choice(persons_by_class[0])
    person_1 = np.random.choice(persons_by_class[1])
    df_0, df_1 = samples.query(f'person_id == {person_0}'), samples.query(f'person_id == {person_1}')
    probs = pd.concat([df_0, df_1])

    fig, axes, _ = sp.compare_cont_dists([probs], ['prob'], groupby='y', nbins=15)
    axes[0][0].set_xlim(0, 1)
    axes[0][0].vlines(df_0.nn_prob.unique(), 0 , 5)
    axes[0][0].vlines(df_1.nn_prob.unique(), 0 , 5)
```

### Bayesian NN with Regularization

#### Model definition 

---------------------------------
We will change the prior for weight parameters $\mathbf{w}$ by adding regularization such that the mean probability for males and females are close.

Let $F$ be the subset of data that correspond to females and $M$ be males we define the fairness of our model to be:

$$
f(\mathbf{w}) = \frac{\sum_{i \in F} P(x_i | \mathbf{w})}{\sum_{i \in M} P(x_i | \mathbf{w})}
$$

We know that for the fastai trained model $f(\mathbf{w}) = 0.35$ and we need to push this value to be $1$. So our new model will be 

$$\begin{align*}  y &\sim \text{Bernoulli}(P(x | \mathbf{w})) \\
\mathbf{w} &\sim \text{Normal}(w^*, \sigma^2) \\
f(\mathbf{w}) &\sim \text{Normal}(1, \lambda^2)
\end{align*} 
$$


With this definition we have that our non-normalized log-posterior density for the weights $\mathbf{w}$ is:

$$
\begin{align*}
L(\mathbf{w})
& = - \left( \sum_{i=1}^N \text{CrossEntropy}(y_i, P(x_i | \mathbf{w})) +  \sum_{j=1}^d  \left[\frac{1}{2\sigma_j^2}(\mathbf{w}_j - w^*_j)^2 + \log \sigma_j \right] + \left[\frac{1}{2\lambda^2}(f(\mathbf{w}) - 1)^2 + \log \lambda \right]\right)
\end{align*} 
$$

--------------------------------

We extract tensors of numerical and categorical variables on the training set

```python
device = learn.dl.device

x_num_train = torch.Tensor(learn.dls.dataset.conts.values).to(device, torch.float32)
x_cat_train = torch.LongTensor(learn.dls.dataset.cats.values).to(device)
y_train = torch.Tensor(learn.dls.dataset.y.values).to(device)
x_num_train, x_cat_train, y_train
```

```python
learn.dls.dataset.cat_names
```

```python
gender_dim = 6
print(x_cat_train[:, gender_dim]) # variable de genero, 2='hombres', 1='mujeres'
males = torch.nonzero(x_cat_train[:, gender_dim] == 2)
females = torch.nonzero(x_cat_train[:, gender_dim] == 1)

y_train[males].mean(), y_train[females].mean()
```

```python
zero = torch.zeros(1, ).to(device)
one = torch.ones(1, ).to(device)
eps = torch.Tensor([1e-10]).to(device)
```

```python
scale = 0.0001

num_iterations = 1000
batch_size = 10000
fastai_centered = True

lmbda_range = np.linspace(0.3, 0.01, 30)
#lmbda_range = np.linspace(10, 1.5, 30)

results_df = pd.DataFrame(columns=['auc', 'fairness', 'lambda'])
results_df['lambda'] = lmbda_range
results_df.set_index('lambda', inplace=True)

for lmbda_regularization in tqdm(lmbda_range):
    def model(x_cat, x_num, y=None):

        #define priors
        priors = {}
        for name, values in learn.model.named_parameters():
            if fastai_centered:
                loc = values
            else:
                loc = torch.zeros_like(values)

            priors[name] = dist.Normal(loc=loc, 
                                       scale=scale*torch.ones_like(values))

        lifted_module = pyro.random_module("module", learn.model, priors)
        lifted_module_sample = lifted_module()

        probs = softmax(lifted_module_sample(x_cat, x_num), dim=1)[:, 1] #probabilidad clase 1
        
        #likelihood:
        if y is not None:
            with pyro.plate('observe_data', size=len(y), subsample_size=batch_size, device=device) as ind:
                pyro.sample('obs', dist.Bernoulli(probs.index_select(0, ind)), obs=y.index_select(0, ind))

        #regularization
        males = torch.nonzero(x_cat[:, gender_dim] == 2)
        females = torch.nonzero(x_cat[:, gender_dim] == 1)
        mean_pred_males = probs[males].mean()      #mean probability for males
        mean_pred_females = probs[females].mean()  #mean probability for females

        #unfairness = torch.pow(mean_pred_males - mean_pred_females, 2)         
        #pyro.sample('regular', dist.Normal(loc=zero, scale=lmbda_regularization), 
        #        obs=unfairness)
            
        #unfairness = mean_pred_males/mean_pred_females-1
        #fairness = torch.clamp(mean_pred_females/(mean_pred_males + eps), 0.001, .999)
        fairness = mean_pred_females/(mean_pred_males + eps)
        #print(fairness)
        #print(mean_pred_females, mean_pred_males)
        pyro.sample('regular', dist.Normal(one, one*lmbda_regularization), 
            obs=fairness)
        #pyro.sample('regular', dist.Beta(2*one, one), 
        #    obs=fairness)

        return probs

    guide = AutoDiagonalNormal(model)
    optim = Adam({"lr": 0.03})
    svi = SVI(model, guide, optim, loss=Trace_ELBO(), num_samples=1000)


    pyro.clear_param_store()
    loss_list = []

    for j in range(num_iterations):
        # calculate the loss and take a gradient step
        loss = svi.step(x_cat_train, x_num_train, y_train)
        loss_list += [loss]
        if j % 100 == 0:
            print("[iteration %04d] loss: %.4f" % (j + 1, loss))

    iter_min = 0
    plt.plot(np.array(loss_list[iter_min:]))
    plt.xlabel('iterations')
    plt.ylabel('loss');
    plt.show()

    df_valid = df.query('is_valid == 1').copy()
    valid_post = get_predictions_pyro(model, guide, df_valid, n_samples=100)['_RETURN']

    df_valid['mean_prob_post'] = valid_post.mean(axis=0)

    y_true = df_valid.income.replace({'<=50K': 0, '>50K': 1}).values
    roc_fastai = roc_auc_score(y_true, df_valid['prob>50K'].values)
    roc_mean_bayesian = roc_auc_score(y_true, df_valid['mean_prob_post'].values)
    print(roc_fastai, roc_mean_bayesian)

    p_male = df_valid.query('gender=="Male"').mean_prob_post.mean()
    p_female = df_valid.query('gender=="Female"').mean_prob_post.mean()
    fairness_bnn = (p_female/p_male)
    print(unfairness_fastai, fairness_bnn)
    results_df.loc[lmbda_regularization, 'auc'] = roc_mean_bayesian
    results_df.loc[lmbda_regularization, 'fairness'] = fairness_bnn
    
    fig, ax = plt.subplots(figsize=(7,7))

    with plt.xkcd():
        sn.scatterplot('auc', 'fairness', s=400, color='b', data=results_df, ax=ax, alpha=0.7)
    plt.show()
    
    results_df.dropna()[['auc']].plot(marker='o')
    plt.show()
    results_df.dropna()[['fairness']].plot(marker='o')
    plt.show()
```
